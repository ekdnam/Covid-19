```python
time = main_df['created_at']
text = main_df['text']
final_df = pd.DataFrame(columns={'text', 'date', 'time'})
```

```python
final_df['text'] = main_df['text']
dates = []
times = []
```

```python
for item in time.iteritems():
    date, time = item[1].split(' ')
    dates.append(date)
    times.append(time)
```

```python
final_df['date'] = dates
final_df['time'] = times
final_df['hashtag'] = ''
final_df['lang'] = 'en'
```
